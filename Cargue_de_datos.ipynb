# **Tipos de archivos para bases de datos**

**XLS (Excel Spreadsheet):**

Es un formato de archivo creado por Microsoft Excel.
Almacena datos en una hoja de cálculo con celdas organizadas en filas y columnas.
Puede contener fórmulas, gráficos y macros.

**XLSX (Excel Open XML Spreadsheet):**

También es un formato de archivo de Excel, pero utiliza un estándar abierto basado en XML.
Permite almacenar más datos que el formato XLS y es más eficiente en cuanto a espacio.

**CSV (Comma-Separated Values):**

Es un formato de archivo de texto que utiliza comas para separar los valores.
Es simple y ampliamente compatible, pero carece de funciones avanzadas de formato.

**PDF (Portable Document Format):**

No es específico de hojas de cálculo, sino un formato de documento independiente del software.
Mantiene el formato original, es de solo lectura y es ampliamente utilizado para compartir documentos.

**Archivos .doc y  .docx**
Es un formato de documento de texto abierto inventado originalmente por Microsoft que se utiliza para crear archivos de texto editables

**Archivos .txt**
Un archivo de texto (txt) es un formato de archivo que almacena información de manera simple y sin formato. Contiene texto sin ningún tipo de formato de fuente, estilo o formato de página.
